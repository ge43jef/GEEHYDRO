```python
import pandas as pd
import numpy as np

file_paths = ['/content/FLX_US-Blo_FLUXNET2015_SUBSET_DD_1997-2007_1-4.csv',
              '/content/FLX_US-Ton_FLUXNET2015_SUBSET_DD_2001-2014_1-4.csv',
              '/content/FLX_US-Tw3_FLUXNET2015_SUBSET_DD_2013-2014_2-4.csv',
              '/content/FLX_US-Twt_FLUXNET2015_SUBSET_DD_2009-2014_1-4.csv',
              '/content/FLX_US-Var_FLUXNET2015_SUBSET_DD_2000-2014_1-4.csv',
              '/content/era5LandBLO_values.csv',
              '/content/era5LandTON_values.csv',
              '/content/era5LandTW3_values.csv',
              '/content/era5LandTWT_values.csv',
              '/content/era5LandVAR_values.csv']

# Initialize an empty list to store the data frames
data_frames = []

# Iterate over the first 5 files and select the desired columns
for file_path in file_paths[:5]:
    data = pd.read_csv(file_path, delimiter=",", skipinitialspace=True, parse_dates=True)
    selected_data = pd.DataFrame(
        {"sw": data.SW_IN_F, "lw": data.LW_IN_F,
         "pre": data.PA_F, "u10": data.WS_F, "vpd": data.VPD_F, "lh": data.LE_CORR})
    data_frames.append(selected_data)

# Concatenate the tables from the first 5 files
first_combined_data = pd.concat(data_frames, axis=0, ignore_index=True)

# Clear the data_frames list
data_frames.clear()

# Initialize a new DataFrame for the remaining files
remaining_data = pd.DataFrame()

# Iterate over the remaining files and select the desired columns
for file_path in file_paths[5:]:
    data = pd.read_csv(file_path, delimiter=",", skipinitialspace=True, parse_dates=True)
    selected_data = pd.DataFrame(
        {"temperature": data.temperature_2m, "precipitation": data.total_precipitation_sum})
    remaining_data = pd.concat([remaining_data, selected_data], ignore_index=True)

# Match lengths using broadcasting
n = len(first_combined_data)
m = len(remaining_data)
combined_data = pd.concat([first_combined_data, remaining_data.iloc[:n, :].reset_index(drop=True)], axis=1)

##combined_data.head(150)
```
