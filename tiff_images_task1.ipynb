```python
import rasterio
import matplotlib.pyplot as plt
%matplotlib inline

# Read the raster file
file_path = "AA1122dd.tif"
with rasterio.open(file_path) as dataset:
    # Read all bands in the raster file
    image = dataset.read()
    # Retrieve the spatial information
    transform = dataset.transform

# Select band 1 and band 2
band1 = image[0]
band2 = image[1]

# Generate x and y coordinates based on the spatial information
rows, cols = band1.shape
x = [transform[2] + i * transform[0] for i in range(cols)]
y = [transform[5] + i * transform[4] for i in range(rows)]

# Plot band 1 and band 2 in the same figure
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(band1, cmap='viridis', extent=[x[0], x[-1], y[-1], y[0]])
plt.title('Precipitation')
plt.colorbar(label='Pixel Value')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.subplot(1, 2, 2)
plt.imshow(band2, cmap='viridis', extent=[x[0], x[-1], y[-1], y[0]])
plt.title('Temperature')
plt.colorbar(label='Pixel Value')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.tight_layout()
plt.show()
```
