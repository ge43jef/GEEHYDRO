```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MaxAbsScaler

from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df =pd.read_csv('/content/task1data.csv')
df.head()
```

```python
plt.scatter(df['mean_precipitation'], df['mean_temperature'])
```

```python
scaler = MinMaxScaler()
scaler.fit(df.mean_precipitation.values.reshape(-1, 1))
df.mean_precipitation = scaler.transform(df.mean_precipitation.values.reshape(-1, 1))

scaler.fit(df.mean_temperature.values.reshape(-1, 1))
df.mean_temperature = scaler.transform(df.mean_temperature.values.reshape(-1, 1))
df
```

```python
km = KMeans(n_clusters=4)
y_predict = km.fit_predict(df[['mean_precipitation',	'mean_temperature']])
y_predict
```

```python
df['cluster'] = y_predict
df
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
df3 = df[df.cluster == 2]
df4 = df[df.cluster == 3]
plt.scatter (df1.mean_precipitation,df1.mean_temperature)
plt.scatter (df2.mean_precipitation,df2.mean_temperature)
plt.scatter (df3.mean_precipitation,df3.mean_temperature)
plt.scatter (df4.mean_precipitation,df4.mean_temperature)
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color ='purple', marker= '*', label='Centroid')
plt.legend()
```

```python
k_rng = range (1, 12)
sse =[]
for k in k_rng:
  km=KMeans(n_clusters=k)
  km.fit(df[['mean_precipitation','mean_temperature']])
  sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```
